```python
# YOUR CODE HERE

plt.xlabel('Amount of CO2')

plt.xticks(range(fuel_econ['co2'].min()-4, fuel_econ['co2'].max()+ 1, 50))
plt.yticks(range(0, 400, 30))
plt.xlim(fuel_econ['co2'].min(), fuel_econ['co2'].max())
bins = np.arange(fuel_econ['co2'].min(), fuel_econ['co2'].max()+5, 20)
# print(bin)
plt.hist(data = fuel_econ, x = 'co2', bins = bins,
         edgecolor='b', alpha=0.8, color='orange')

plt.title('CO2 distribution')
# print(plt.ylabel)
print(fuel_econ['co2'].max())
print(fuel_econ['co2'].min())
plt.show()
```

DEFINITION: The distribution of the values in the CO2 column is symmetric.

```python
# YOUR CODE HERE

plt.figure(figsize=(10, 6))
plt.hist(fuel_econ['highway'], bins=30, edgecolor='black', color='purple')
plt.xlim(fuel_econ['highway'].min(), fuel_econ['highway'].max(), 5)
plt.title('Highway Distribution')
plt.xlabel('Cylinders Values')
plt.ylabel('Frequency')
plt.show()
```

```python
# YOUR CODE HERE

plt.figure(figsize=(10, 6))
plt.hist(fuel_econ['UCity'], bins=50, edgecolor='black', color='lightgreen')
plt.xlim(fuel_econ['UCity'].min(), fuel_econ['UCity'].max(), 5)
plt.title('UCity Distribution')
plt.xlabel('UCity Values')
plt.ylabel('Frequency')
plt.show()
```

**TO DO:** create a heat map of engine displacement i.e the column "disp"  vs. co2 production.and interpret it (is there correlation?what did you notice?). Engine Displacement is in litres

```python
# Creating a heatmap
plt.figure(figsize=(10, 6))
sb.heatmap(data=fuel_econ[['displ', 'co2']].corr(), annot=True, cmap='YlGnBu', fmt=".2f")
plt.xlabel('Engine Displacement in (L)')
plt.ylabel('CO2 Production in (g/mi)')
plt.title('Correlation Heatmap: Engine Displacement vs. CO2 Production')
plt.show()
```
